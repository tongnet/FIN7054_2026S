```python
# firm value
cf = [11, 12.1]
v = 0
r = 0.1
i=1
for i in range(len(cf)):
    v += cf[i]/((1+r)**(i+1))
v = round(v,2)
print(f"the firm value is {v} million dollars")
```

```python
# Firm value without insurance
cf1 = {300:0.8, 200:0.2}
expected_cf1 = 0
for cf, p in cf1.items():
    expected_cf1 += p*cf
V = round(expected_cf1/1.1,2)
print(f"the firm value is {V} million dollars")
```

```python
# estimate insurance cost (loading)
loss = {0:0.8, 100:0.2}
expected_loss = 0
for loss1, p in loss.items():
    expected_loss += p*loss1
fair_premium = round(expected_loss/1.1,2)
print(f"the actuarially fair premium is {fair_premium} million dollars")
loading = round(20-fair_premium, 2)
print(f"Loading is {loading} million dollars")
```

```python
# calculate tax payments
def tax(amount):
    if 0<= amount <= 3:
        return amount*0.1
    elif 3<amount<=7:
        return 3*0.1+(amount-3)*0.2
    elif 7<amount<=12:
        return 3*0.1+(7-3)*0.2+(amount-7)*0.3
    else:
        print("no a valid amount")

income = {2:0.5, 10:0.5}
expected_income = 0
expected_tax = 0
for v, p in income.items():
    expected_income += p*v
    expected_tax += p*tax(v)
print(f"expected tax without hedging is {expected_tax} million dollars")
print(f"expected after-tax income without hedging is {expected_income-expected_tax} million dollars")
tax_hedged = round(tax(expected_income),4)
print(f"total tax with hedging is {tax_hedged} million dollars")
tax_savings = round(expected_tax - tax_hedged, 4)
print(f"tax savings due to risk management is {tax_savings} million dollars")
print(f"expected after-tax income with hedging is {expected_income-tax_hedged} million dollars")
```
