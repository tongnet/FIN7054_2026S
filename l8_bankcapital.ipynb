```python
# Bank capital ratios

# Input data for capital ratios
T1 = float(input("What's Tier I capital (in $m)?\n"))
CAP = float(input("What's total risk adjusted capital (in $m)?\n"))
RWA = float(input("What's credit risk weighted assets (in $m)?\n"))
TA = float(input("What's total assets (in $m)?\n"))
# Calculate capital and leverage ratios
R1 = T1/RWA
R2 = CAP/RWA
R3 = T1/TA
print(f"The bank’s Tier I capital ratio is {R1}")
print(f"The bank’s total risk-based capital ratio is {R2}")
print(f"The bank’s Tier I leverage ratio is {R3}")
```
