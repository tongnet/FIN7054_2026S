```python
import numpy as np
import scipy
from scipy.stats import norm
DEAR_BOND = 1000000*(7/1.07243)*0.001* norm.ppf(0.99)
DEAR_FX = 800000*1.25* norm.ppf(0.99)*0.00565
DEAR_STOCK = 10000000* norm.ppf(0.99)*0.002
print("DEAR for the bond is",DEAR_BOND)
print("DEAR for the Euro is",DEAR_FX)
print("DEAR for the stock is", DEAR_STOCK)

corr12=-0.2
corr13=0.4
corr23=0.1

DEAR = DEAR_BOND**2 + DEAR_FX**2 + DEAR_STOCK**2+ 2*corr12*DEAR_BOND*DEAR_FX + 2*corr13*DEAR_BOND*DEAR_STOCK + 2*corr23*DEAR_FX*DEAR_STOCK

DEAR_P = np.sqrt(DEAR)
print("DEAR for the portfolio is", DEAR_P)
```

```python
import numpy as np
import scipy
from scipy.stats import norm
DEAR_BOND = 1000000*(7/1.07243)*0.001* norm.ppf(0.99)
DEAR_FX = 800000*1.25* norm.ppf(0.99)*0.00565
DEAR_STOCK = 10000000* norm.ppf(0.99)*0.002
print("DEAR for the stock is", DEAR_STOCK)
print("DEAR for the bond is",DEAR_BOND)
print("DEAR for the Euro is",DEAR_FX)
DEAR = np.matrix([[DEAR_STOCK,
                DEAR_BOND,
                DEAR_FX]])

CORR = np.matrix([[1, 0.4, 0.1],
                   [0.4, 1, -0.2],
                   [0.1, -0.2, 1]])

DEAR_P = np.sqrt(DEAR * CORR * DEAR.T)
print("DEAR for the portfolio is", DEAR_P.item((0,0)))
```

```python
# import numpy and generate two numpy arrays
import numpy as np
ret_mkt = np.zeros((10000,)) # 1
ret_p = np.zeros((10000,))

for i in range(len(ret_mkt)): # 2
    ret_mkt[i] = np.random.normal(0.0002, 0.005, 1)[0] # 3
    ret_p[i] = 0.0001 + 1.2 * ret_mkt[i] # 4

# Find the one percentile of the value of ret_p
one_pct = np.percentile(ret_p, 1) # 5
print("the VaR on the portfolio is", one_pct)
```
