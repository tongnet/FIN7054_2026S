```python
# get contract price from polymarket.com
import requests
import json

# 1) Search the event
r = requests.get(
    "https://gamma-api.polymarket.com/public-search",
    params={"q": "US Iran ceasefire", "limit_per_type": 5}
)
data = r.json()

# 2) Take the first matching event
event = data["events"][0]
print("Event:", event["title"])
print()

# 3) Print each market question and its current YES/NO prices
for m in event["markets"]:
    prices = json.loads(m["outcomePrices"])   # e.g. ["0.07","0.93"]
    print(m["question"])
    print("YES:", prices[0], "NO:", prices[1])
    print()
```

```python
# Business risk (flower shop no risk) example
import math
returns = {1:0.5, -0.2:0.5}
mu = 0
var = 0
std = 0
for val, p in returns.items():
    mu += val*p
print("The expected rate of return is", mu)
for val, p in returns.items():
    var += p*(val - mu)**2
std = round(math.sqrt(var),2)
print("The business risk is", std)
```

```python
# financial risk (flower shop using leverage) example
import math
ret_equity = {1.2375:0.5, -0.2625:0.5}
mu_equity = 0
var_equity = 0
std_equity = 0
for val, p in ret_equity.items():
    mu_equity += val*p
print("The expected rate of to the equity holder is", round(mu_equity,4))
for val, p in ret_equity.items():
    var_equity += p*(val - mu_equity)**2
std_equity = round(math.sqrt(var_equity),4)
print("The risk to the business owner is", std_equity)
```

```python
# Expected utility using 1 boat versus using two boats
import math
x = {4000:0.5, 12000:0.5}    # distribution of 1 boat: x
y = {4000:0.25, 8000:0.5, 12000:0.25}    # distribution of 2 boats: y
ux = 0
for vx, px in x.items():
    ux += px*math.sqrt(vx)
uy = 0
for vy, py in y.items():
    uy += py*math.sqrt(vy)
print("The expected utility of x is", round(ux,2))
print("The expected utility of y is", round(uy,2))
```

```python
# expected utility and risk premium under normal condition
import math
x = {12000:0.5, 4000:0.5}
u = 0
ex=0
for val, p in x.items():
    u += p*math.sqrt(val)
    ex += p*val
ce=u**2
pi = ex - ce
print("The expected utility is", round(u,2))
print("The certainty equivalent is", round(ce,2))
print("The risk premium is", round(pi,2))
```

```python
# expected utility and risk premium under worst case scenario

import math
x = {12000:0.1, 4000:0.9}
u = 0
ex=0
for val, p in x.items():
    u += p*math.sqrt(val)
    ex += p*val
ce=u**2
pi = ex - ce
print("The expected utility is", round(u,2))
print("The certainty equivalent is", round(ce,2))
print("The risk premium is", round(pi,2))
```
