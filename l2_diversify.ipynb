```python
import math
x = [9, 4, 10, 5]
y = [26, 30, 20, 24]
p = [0.2, 0.3, 0.3, 0.2]
mu_x = 0
var_x = 0
mu_y = 0
var_y = 0
covar = 0
for i in range(len(x)):
    mu_x += p[i]*x[i]
    mu_y += p[i]*y[i]
print("the expected value of X is", mu_x)
print("the expected value of Y is", mu_y)
for i in range(len(x)):
    var_x += p[i]*(x[i] - mu_x)**2
    var_y += p[i]*(y[i] - mu_y)**2
    covar += p[i]*(x[i] - mu_x)*(y[i] - mu_y)
print("the standard deviation of X is", math.sqrt(var_x))
print("the standard deviation of Y is", math.sqrt(var_y))
print("the covariance between X and Y is", covar)
corr = covar/math.sqrt(var_x*var_y)
print("the correlation coefficient between X and Y is", corr)
```

```python
!pip install numpy
!pip install matplotlib

import numpy as np
import matplotlib.pyplot as plt
import math

mu_a = 0.1
mu_b = 0.2
std_a = 0.15
std_b = 0.3
weight_a = 0.4
weight_b = 0.6

rho = np.linspace(-1,1,1000)
mu_p = np.zeros((1000,))
std_p = np.zeros((1000,))

for i in range(1000):
    mu_p[i] = weight_a*mu_a + weight_b*mu_b
    std_p[i] = math.sqrt(weight_a**2*std_a**2 + weight_b**2*std_b**2 + 2*weight_a*weight_b*rho[i]*std_a*std_b)

plt.plot(rho, mu_p, color='blue', label='portfolio return')
plt.plot(rho, std_p, color='green',label='portfolio volatility', linestyle='--')
plt.xlabel('correlation coefficient between stocks')
plt.legend()
plt.show()
```
